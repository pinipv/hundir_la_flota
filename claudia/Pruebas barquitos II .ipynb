```python
import random
import numpy as np
```

```python
#1 Constantes del juego


#Mensajes
Mensaje_inicial = "Hola! bienvenido al juego"
Instrucciones = "Instrucciones"
Fallo = "Has Fallado"
Perder_juego = "Has Perdido"
Fin = "Fin de partida"


#Barcos
"""Diccionario donde la clave es el barco y el valor es la eslora"""

Diccionario_barcos = {"B1S1": 1, "B2S1": 1, "B3S1": 1, "B4S1": 1, "B1S2": 2, "B2S2": 2, "B3S2": 2, "B1S3": 3, "B2S3": 3, "B1S4":4 }


#Impactos
Agua = " "
Disparo = "B"
Disparo_fallido = "-"



class Tablero:
    Id_jugador = " "
    Filas = 10
    Columnas = 10
    Barcos = Diccionario_barcos
    Tablero_propio = []
    Tablero_contrincante = []


    def __init__(self,Id_jugador):
        self.Id_jugador = Id_jugador
        self.Tablero_propio = np.full(fill_value = " ", shape = (self.Filas, self.Columnas))
        self.Tablero_contrincante = np.full(fill_value = " ", shape = (self.Filas, self.Columnas))



    def mostrar_tablero(self):
        print(self.Id_jugador)
        print(self.Tablero_propio)
    


    def posicionar_barco(self, eslora):
        while True:
            orientacion = random.choice(['N', 'S', 'E', 'O'])

            # Posicion inicial del barco
            posicion_actual = np.random.randint(10, size = 2)
            
            fila = posicion_actual[0]
            col = posicion_actual[1]
            
            # Recogemos las 4 posiciones colindantes a posicion_actual
            colindantes_Norte = self.Tablero_propio[fila:fila - eslora:-1, col]
            colindantes_Este = self.Tablero_propio[fila, col: col + eslora]
            colindantes_Sur = self.Tablero_propio[fila:fila + eslora, col]
            colindantes_Oeste = self.Tablero_propio[fila, col: col - eslora:-1]
            

            # Comprobar si esas posiciones son validas
            # Orientacion Norte
            if orientacion == 'N' and 0 <= fila - eslora < 10 and 'O' not in colindantes_Norte:
                self.Tablero_propio[fila:fila - eslora:-1, col] = 'O'
                break

            # Orientacion Este
            elif orientacion == 'E' and 0 <= col + eslora < 10 and 'O' not in colindantes_Este:
                self.Tablero_propio[fila, col: col + eslora] = 'O'
                break

            # Orientacion Sur
            elif orientacion == 'S' and 0 <= fila + eslora < 10 and 'O' not in colindantes_Sur:
                self.Tablero_propio[fila:fila + eslora, col] = 'O'
                break

            # Orientacion Oeste
            elif orientacion == 'O' and 0 <= (col - eslora+1) < 10 and 'O' not in colindantes_Oeste:
                if ((col-eslora)== -1):
                    self.Tablero_propio[fila, col::-1] = 'O'           
                else:
                    self.Tablero_propio[fila, col: (col - eslora):-1] = 'O'
                break

            else:
                continue
        
    def aniadir_barcos_tablero(self):
        for clave,valor in self.Barcos.items():
            self.posicionar_barco(valor)



        





#Programa Principal

print(Mensaje_inicial)

tablero_jugador = Tablero("Claudia")  #pensando que aqui pediremos un input para que de el nombre de usuario.Esto es temporal para ir probando
tablero_jugador.mostrar_tablero()
tablero_jugador.aniadir_barcos_tablero()
tablero_jugador.mostrar_tablero()


tablero_maquina = Tablero("Maquina")
tablero_maquina.mostrar_tablero()
tablero_maquina.aniadir_barcos_tablero()
tablero_maquina.mostrar_tablero()
    
```
